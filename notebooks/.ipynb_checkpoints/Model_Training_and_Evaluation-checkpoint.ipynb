# Model Training and Evaluation - Vehicle Claim Fraud DetectionThis notebook trains an XGBoost classifier on the vehicle claim fraud dataset and evaluates its performance. The trained model is then saved to disk.

```python
import sys
import os
sys.path.append(os.path.abspath('..'))

import warnings
warnings.filterwarnings('ignore')
```

```python
import os
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import xgboost as xgb
import joblib
from src.feature_engineering import create_features_and_target
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, precision_recall_curve

params = {
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.05, 0.1],
    'n_estimators': [100, 200],
    'subsample': [0.7, 0.8, 1],
}

# Load dataset
data_path = os.path.join('..', 'data', 'fraud_oracle.csv')
data = pd.read_csv(data_path)
print('Dataset shape:', data.shape)

# Create features and target
features, target = create_features_and_target(data)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(
    features, target, test_size=0.2, random_state=42, stratify=target
)

# Calculate original class ratio
original_neg_count = sum(y_train == 0)
original_pos_count = sum(y_train == 1)
original_scale_pos_weight = original_neg_count / original_pos_count

print(f"Scale_pos_weight before SMOTE: {original_scale_pos_weight:.2f}")

# Apply SMOTE to balance classes
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

# Confirm class balancing after SMOTE
res_neg_count = sum(y_train_res == 0)
res_pos_count = sum(y_train_res == 1)
res_scale_pos_weight = res_neg_count / res_pos_count

print(f"Scale_pos_weight after SMOTE: {res_scale_pos_weight:.2f}")

# Train an XGBoost classifier without scale_pos_weight (already balanced by SMOTE)
#model = xgb.XGBClassifier(
#    use_label_encoder=False,
#    eval_metric='logloss',
#    random_state=42
#)
#model.fit(X_train_res, y_train_res)
grid = GridSearchCV(
    xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss'),
    param_grid=params,
    scoring='f1_macro',
    cv=5,
    verbose=1
)

grid.fit(X_train_res, y_train_res)
print("Best parameters:", grid.best_params_)

# Evaluate the model
y_probs = grid.predict_proba(X_test)[:, 1]

precision, recall, thresholds = precision_recall_curve(y_test, y_probs)

# Ensure the lengths match before calculating best threshold
best_threshold = thresholds[np.argmax(precision[:-1] - recall[:-1])]
print("best_threshold:", best_threshold)
best_threshold = 0.5

y_pred_adjusted = (y_probs > best_threshold).astype(int)

print('Classification Report:')
print(classification_report(y_test, y_pred_adjusted))

# Save the model
model_save_path = os.path.join('..', 'model', 'model_xgb.pkl')
os.makedirs(os.path.dirname(model_save_path), exist_ok=True)
joblib.dump(grid, model_save_path)
print('Model saved as model_xgb.pkl')
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, roc_curve, auc
import seaborn as sns
import shap

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred_adjusted)
plt.figure(figsize=(6,  4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# ROC Curve
probs = grid.predict_proba(X_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, probs)
roc_auc = auc(fpr, tpr)
plt.figure(figsize=(6, 4))
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc='lower right')
plt.show()

# SHAP Analysis
# Create a TreeExplainer for the model
# Ensure SHAP does not show warnings
shap.initjs()

# Create a TreeExplainer for the **best model**, not GridSearchCV
explainer = shap.TreeExplainer(grid.best_estimator_)
shap_values = explainer.shap_values(X_test)

# Visualize SHAP Summary Plot
shap.summary_plot(shap_values, X_test)
```
